```python
import os
import torch
from torch.nn.functional import cosine_similarity

PROJECT_ROOT = os.path.dirname(os.getcwd())
CACHE_DIR_NAME = "activation_cache"

# Vector1: singolo file da hallOG.py (belief_bank)
file_path1 = os.path.join(PROJECT_ROOT, CACHE_DIR_NAME, "Qwen2.5-7B", "belief_bank", "activation_attn", "layer0_activations.pt")
vector1 = torch.load(file_path1)[0]

# Vector2: dal batch combinato (belief_bank1)
file_path2 = os.path.join(PROJECT_ROOT, CACHE_DIR_NAME, "Qwen2.5-7B", "belief_bank1", "activation_attn", "layer0_activations.pt")
all_vectors = torch.load(file_path2)
vector2 = all_vectors[0]

print(f"\nCosine similarity: {cosine_similarity(vector1.float(), vector2.float(), dim=0):.4f}")

```
